```python
import torch

# A tensor we want gradients for
x = torch.tensor(3.0, requires_grad=True)

# Build a tiny computation
y = x ** 2      
z = 2 * y + 1   

# Walk the graph backward from z
z.backward()

print("x:", x.item())
print("dz/dx:", x.grad.item())
```

```python
import torch
import torch.nn as nn

X = torch.tensor([[0,0],[0,1],[1,0],[1,1]], dtype=torch.float32)
Y = torch.tensor([[0],[1],[1],[0]], dtype=torch.float32)

class XORNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(2, 4)
        self.fc2 = nn.Linear(4, 1)

    def forward(self, x):
        # apply fc1, then sigmoid, then fc2, then sigmoid
        # use torch.sigmoid(...)
        z1 = self.fc1(x)
        a1 = torch.sigmoid(z1)
        z2 = self.fc2(a1)
        a2 = torch.sigmoid(z2)
        
        return a2

net = XORNet()
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(net.parameters(), lr=0.5)

for epoch in range(5000):
    optimizer.zero_grad()
    output = net(X)  
    loss = criterion(output, Y)
    loss.backward()
    optimizer.step()

    if epoch % 500 == 0:
        print(f"Epoch {epoch:4d} — loss: {loss.item():.6f}")

print("\nPredictions:")
print(net(X).detach().round())
print("Targets:")
print(Y)
```
